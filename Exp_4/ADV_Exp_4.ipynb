```r
# Load necessary libraries
library(ggplot2)
library(dplyr)

# Read the data
crime_data <- read.csv("/content/police_shootings.csv")
```

```r
# Bar Chart for Manner of Death
ggplot(crime_data, aes(x=manner_of_death)) +
  geom_bar(fill="steelblue") +
  labs(title="Frequency of Manner of Death", x="Manner of Death", y="Count")
```

```r
# Pie Chart for Armed Status
armed_data <- crime_data %>%
  count(armed) %>%
  mutate(percentage = n / sum(n) * 100)

ggplot(armed_data, aes(x="", y=percentage, fill=armed)) +
  geom_bar(width = 1, stat = "identity") +
  coord_polar(theta = "y") +
  labs(title="Breakdown by Armed Status")
```

```r
# Histogram for Age Distribution
ggplot(crime_data, aes(x=age)) +
  geom_histogram(bins=30, fill="tomato", color="black") +
  labs(title="Age Distribution of Victims", x="Age", y="Frequency")
```

```r
# Timeline Chart
crime_data$date <- as.Date(crime_data$date) # Ensure 'date' is in Date format

ggplot(crime_data, aes(x=date)) +
  geom_line(stat="count", color="green") +
  labs(title="Incidents Over Time", x="Time", y="Number of Incidents")
```

```r
# Scatter Plot for Age vs Threat Level
ggplot(crime_data, aes(x=age, y=threat_level)) +
  geom_point(aes(color=threat_level), alpha=0.6) +
  labs(title="Age vs. Threat Level", x="Age", y="Threat Level")
```

```r
# Bubble Plot for Geographic Distribution
ggplot(crime_data, aes(x=longitude, y=latitude, size=age)) +
  geom_point(alpha=0.6) +
  labs(title="Geographic Distribution of Incidents", x="Longitude", y="Latitude")
```
